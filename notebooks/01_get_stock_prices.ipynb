```python
import yfinance as yf
import pandas as pd
import numpy as np
from openpyxl import load_workbook
import openpyxl
```

```python
# Load stock tickers from Excel file
tickers = pd.read_excel('Data/tickers.xlsx')

# Initialize an empty DataFrame to store all downloaded data
all_prices = pd.DataFrame()
```

```python
# Read company names from the 'ticker' column in the Excel file
company_names = []
for index, row in tickers.iterrows():
    company_names.append(row['ticker'])
```

```python
# Download stock price data using yfinance
prices = yf.download(company_names, start="2000-01-01", end=None, interval="1d", group_by='ticker')

# Append downloaded data to the main DataFrame
all_prices = pd.concat([all_prices, prices], axis=0)
```

```python
# Save price data to a new Excel sheet
new_wb = openpyxl.Workbook()
new_ws = new_wb.active

# Write header (Date + company names)
new_ws.append(['Date'] + company_names)

# Write rows with closing prices
for i in range(len(all_prices)):
    date = all_prices.index[i].strftime("%Y-%m-%d")
    prices_list = [
        date
    ] + [
        all_prices.iloc[i][company]['Close']
        if not pd.isna(all_prices.iloc[i][company]['Close'])
        else None
        for company in company_names
    ]
    new_ws.append(prices_list)
```

```python
# Save the Excel file
new_file_name = "Data/all_stock_prices.xlsx"
new_wb.save(new_file_name)
```
