```python
import yfinance as yf
import pandas as pd
import numpy as np
from openpyxl import load_workbook
import openpyxl
```

```python
# Charger les noms des actions à partir du fichier Excel
tickers = pd.read_excel('Data/tickers.xlsx')

# Initialiser un DataFrame vide pour stocker toutes les données téléchargées
all_prices = pd.DataFrame()
```

```python
# Lecture des noms de sociétés à partir de la colonne A du fichier Excel
company_names = []
for index, row in tickers.iterrows():
    company_names.append(row['ticker'])
```

```python
# Téléchargement des données de prix à l'aide de la bibliothèque yfinance
prices = yf.download(company_names, start="2000-01-01", end=None, interval="1d", group_by='ticker')
all_prices = pd.concat([all_prices, prices], axis=0)
```

```python
# Sauvegarde des données de prix dans une nouvelle feuille Excel
new_wb = openpyxl.Workbook()  # Create a new Excel workbook
new_ws = new_wb.active
new_ws.append(['Date'] + company_names)
for i in range(len(all_prices)):
    date = all_prices.index[i].strftime("%Y-%m-%d")
    prices_list = [date] + [all_prices.iloc[i][company]['Close'] if not pd.isna(all_prices.iloc[i][company]['Close']) else None for company in company_names]
    new_ws.append(prices_list)
```

```python
# Sauvegarde du fichier Excel
new_file_name = "Data/all_stock_prices.xlsx"  # Provide a different file name for the new Excel file
new_wb.save(new_file_name)
```
