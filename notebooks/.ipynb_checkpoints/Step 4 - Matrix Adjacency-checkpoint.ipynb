```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
# Fonction pour traiter un DataFrame
def process_dataframe(df):
    # Seuils
    p_value_threshold_1 = 0.05
    p_value_threshold_2 = 1

    # Parcourir chaque élément de la matrice
    for i in range(df.shape[0]):
        for j in range(df.shape[1]):
            if i != j:  # Garder 1 pour les interactions entre les sociétés
                if df.iloc[i, j] <= p_value_threshold_1:
                    df.iloc[i, j] = 1
                elif df.iloc[i, j] < p_value_threshold_2:
                    df.iloc[i, j] = 0

    return df
```

```python
# Dossier d'origine
input_folder = "Data/PreparedMatrixAdjacency/"

# Dossier de destination
output_folder = "Data/MatrixAdjacency/"
image_folder = "Image/MatrixAdjacency/"
```

```python
# Vérifier et créer les dossiers de sortie s'ils n'existent pas
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

if not os.path.exists(image_folder):
    os.makedirs(image_folder)
```

```python
# Lire chaque fichier Excel dans le dossier d'origine
for filename in os.listdir(input_folder):
    if filename.endswith(".xlsx"):
        print(f"Processing {filename}...")

        # Lecture du fichier
        df = pd.read_excel(os.path.join(input_folder, filename), header=0, index_col=0)
        
        # Traitement du DataFrame
        gct_df = process_dataframe(df)

        # Enregistrement du DataFrame traité dans le dossier de destination
        gct_df.to_excel(os.path.join(output_folder, "rectified_" + filename))

        # Visualisation et sauvegarde de l'image
        plt.figure(figsize=(18, 12))
        sns.heatmap(gct_df)
        plt.tight_layout()
        plt.savefig(os.path.join(image_folder, filename.replace(".xlsx", ".png")), dpi=300)
        plt.close()  # Fermer la figure après avoir sauvegardé pour éviter d'afficher toutes les figures

print("Finished processing all files!")
```
