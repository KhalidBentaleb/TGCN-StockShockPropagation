```python
from statsmodels.tsa.stattools import grangercausalitytests
import pandas as pd
import numpy as np
import os
```

```python
# Dossier contenant les fichiers Excel
data_folder = "Data/FilteredData"
output_folder = "Data/PreparedMatrixAdjacency"
test = 'ssr_chi2test'
maxlag = 1
```

```python
# Créer le dossier de sortie s'il n'existe pas
os.makedirs(output_folder, exist_ok=True)
```

```python
def grangers_causation_matrix(data, variables, test='ssr_chi2test', maxlag=1, verbose=False):
    """Check Granger Causality of all possible combinations of the Time series.
    The rows are the response variable, columns are predictors. The values in the table
    are the P-Values. P-Values lesser than the significance level (0.05), implies
    the Null Hypothesis that the coefficients of the corresponding past values is
    zero, that is, the X does not cause Y can be rejected.

    data      : pandas dataframe containing the time series variables
    variables : list containing names of the time series variables.
    test      : which test statistic to use for the Granger causality.
                Default is 'ssr_chi2test'.
    maxlag    : the maximum lag to include in the test. Default is 1.
    verbose   : whether to print the Granger Causality and P Values. Default is False.
    """
    df = pd.DataFrame(np.zeros((len(variables), len(variables))), columns=variables, index=variables)
    for c in df.columns:
        for r in df.index:
            try:
                test_result = grangercausalitytests(data[[r, c]], maxlag=maxlag, verbose=False)
                p_values = [round(test_result[i+1][0][test][1], 4) for i in range(maxlag)]
                #if verbose:
                   # print(f'Granger Causality: {r} -> {c}')
                    #print(f'P Values: {p_values}')
                min_p_value = np.min(p_values)
                df.loc[r, c] = min_p_value
            except Exception:
                # Si une exception se produit (y compris une exception non spécifiée),
                # remplir la cellule avec 0
                df.loc[r, c] = 0
    df.columns = [var for var in variables]
    df.index = [var for var in variables]
    return df
```

```python
# Parcourir les fichiers Excel dans le dossier
for filename in os.listdir(data_folder):
    if filename.endswith(".xlsx"):
        file_path = os.path.join(data_folder, filename)
        
        # Extraire la date du nom du fichier (assurez-vous que le nom du fichier contient la date)
        date_str = os.path.splitext(filename)[0].split('_')[-1]
        
        # Charger le fichier Excel
        df = pd.read_excel(file_path)
        
        # Supprimer les colonnes qui ne contiennent pas de données (valeurs nulles)
        df = df.dropna(axis=1, how='all')
        
        # Exclure les colonnes de dates
        columns_to_exclude = ['Date']
        df = df.drop(columns_to_exclude, axis=1)
        
        # Exclure les colonnes avec des valeurs constantes
        constant_columns = [col for col in df.columns if df[col].nunique() == 1]
        df = df.drop(constant_columns, axis=1)
        
        # Calculer la matrice de causalité de Granger
        gc_matrix = grangers_causation_matrix(df, variables=df.columns, verbose=True)
        
        # Chemin et nom du fichier Excel de sortie
        output_file = os.path.join(output_folder, f"matrix_{date_str}.xlsx")
        
        # Stocker le DataFrame dans un fichier Excel
        gc_matrix.to_excel(output_file, index=True)
        
        # Afficher un message de suivi
        print(f"Matrice de causalité de Granger pour {filename} créée avec succès.")
```
