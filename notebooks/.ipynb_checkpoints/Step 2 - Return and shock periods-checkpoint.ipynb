```python
import yfinance as yf
import pandas as pd
import numpy as np
from openpyxl import load_workbook
import openpyxl
```

```python
# Charger les données de tout les cours boursiers
all_stock_prices = pd.read_excel('Data/all_stock_prices.xlsx')

# Initialiser un DataFrame vide pour stocker toutes les données téléchargées
all_prices = pd.DataFrame()
```

```python
print(all_stock_prices.tail(5))
```

```python
# Exclude the "Date" column from the calculation of returns
columns_to_exclude = ['Date']
columns_to_calculate = [col for col in all_stock_prices.columns if col not in columns_to_exclude]

# Calculate logarithmic returns for each column
returns_stock_prices = np.log(all_stock_prices[columns_to_calculate] / all_stock_prices[columns_to_calculate].shift(1))
```

```python
# Ajouter la colonne "Date" au DataFrame des rendements
returns_stock_prices['Date'] = all_stock_prices['Date']

# Réorganiser les colonnes du DataFrame des rendements
returns_stock_prices = returns_stock_prices[['Date'] + columns_to_calculate]

#Remplacer les missing values par zéro
returns_stock_prices = returns_stock_prices.fillna(0)

# Afficher le DataFrame des rendements
print(returns_stock_prices)
```

```python
# Chemin et nom du fichier Excel de sortie
output_file = 'Data/returns_stock_prices.xlsx'

# Stocker le DataFrame dans un fichier Excel
returns_stock_prices.to_excel(output_file, index=False)
```

```python
# Charger les plages de dates à partir du fichier Excel
dates = pd.read_excel('Data/shock_periods.xlsx')

returns_stock_prices['Date'] = pd.to_datetime(returns_stock_prices['Date'])

# Initialiser un dictionnaire vide pour stocker les données téléchargées par plage de dates
filtered_stock_prices_by_date = {}
```

```python
# Télécharger les cours pour chaque action pour chaque plage de dates et stocker les données dans le dictionnaire
for index, date_row in dates.iterrows():
    start_date = pd.to_datetime(date_row['start'])
    end_date = pd.to_datetime(date_row['end'])
    if date_row['number of days'] >= 5:
        new_df = returns_stock_prices[returns_stock_prices['Date'].between(start_date, end_date)]
        filtered_stock_prices_by_date[(start_date, end_date)] = new_df
```

```python
# Trier le dictionnaire par dates
sorted_filtered_prices = sorted(filtered_stock_prices_by_date.items(), key=lambda x: x[0])
```

```python
import os
# Enregistrer les fichiers Excel en ajoutant un numéro de série à la fin du nom
output_directory = 'Data/FilteredData'
if not os.path.exists(output_directory):
    os.makedirs(output_directory)
```

```python
for i, ((start_date, end_date), df) in enumerate(sorted_filtered_prices, start=1):
    file_name = f"{output_directory}/stock_price_{start_date.date()}_{end_date.date()}.xlsx"
    print(f"Saving {file_name}")
    df = df.sort_values('Date')
    df.to_excel(file_name, index=False)
```
