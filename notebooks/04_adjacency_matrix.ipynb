```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
# Function to process a DataFrame
def process_dataframe(df):
    # Thresholds
    p_value_threshold_1 = 0.05
    p_value_threshold_2 = 1

    # Iterate over each element of the matrix
    for i in range(df.shape[0]):
        for j in range(df.shape[1]):
            if i != j:  # Keep interactions between companies
                if df.iloc[i, j] <= p_value_threshold_1:
                    df.iloc[i, j] = 1
                elif df.iloc[i, j] < p_value_threshold_2:
                    df.iloc[i, j] = 0

    return df
```

```python
# Input folder
input_folder = "Data/PreparedMatrixAdjacency/"

# Output folders
output_folder = "Data/MatrixAdjacency/"
image_folder = "Image/MatrixAdjacency/"
```

```python
# Create output folders if they do not exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

if not os.path.exists(image_folder):
    os.makedirs(image_folder)
```

```python
# Read each Excel file from the input folder
for filename in os.listdir(input_folder):
    if filename.endswith(".xlsx"):
        print(f"Processing {filename}...")

        # Load the file
        df = pd.read_excel(os.path.join(input_folder, filename), header=0, index_col=0)
        
        # Process the DataFrame
        gct_df = process_dataframe(df)

        # Save processed DataFrame to output folder
        gct_df.to_excel(os.path.join(output_folder, "rectified_" + filename))

        # Generate and save heatmap image
        plt.figure(figsize=(18, 12))
        sns.heatmap(gct_df)
        plt.tight_layout()
        plt.savefig(
            os.path.join(image_folder, filename.replace(".xlsx", ".png")),
            dpi=300
        )
        plt.close()

print("Finished processing all files!")
```
